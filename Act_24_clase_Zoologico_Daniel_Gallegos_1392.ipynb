### 1. Descripción del Problema

Se busca crear una clase que funcione como un molde maestro, definiendo características comunes (atributos como nombre y capacidad) y comportamientos (funciones como alimentar o guiar tours). A partir de este molde, se deben generar dos instancias u objetos específicos, cada uno con su propia identidad y datos, permitiendo interactuar con ellos de forma independiente.

![texto del vínculo](https://raw.githubusercontent.com/DanielGallegos128/Act24-clase-Zool-gico-Daniel-1392/refs/heads/main/InfoClaseZoologico.png.png)

```python
class Zoologico:
    def __init__(self, nombre, areaTotal, numeroEspecies, horarioApertura):
        # Atributos asignados por el usuario
        self.nombre = nombre
        self.areaTotal = areaTotal
        self.numeroEspecies = numeroEspecies
        self.horarioApertura = horarioApertura

    # Funciones con lógica predeterminada
    def alimentarAnimales(self):
        print(f" > [{self.nombre}]: El equipo de veterinaria está alimentando a las {self.numeroEspecies} especies.")

    def guiarTour(self):
        print(f" > [{self.nombre}]: Iniciando recorrido guiado por las {self.areaTotal} del parque.")

    def mantenerHabitat(self):
        print(f" > [{self.nombre}]: Limpieza de hábitats activa. Recordamos el cierre a las {self.horarioApertura}.")

# --- ENTRADA DE DATOS (INTERACTIVO) ---

def capturar_datos(numero_zoo):
    print(f"\n--- CONFIGURACIÓN DEL ZOOLÓGICO {numero_zoo} ---")
    nom = input("Nombre del zoológico: ")
    area = input("Área total (ej. 50 hectáreas): ")
    especies = input("Número de especies: ")
    horario = input("Horario de apertura (ej. 09:00 - 18:00): ")
    return Zoologico(nom, area, especies, horario)

# Crear los 2 objetos pidiendo datos al usuario
zoologico1 = capturar_datos(1)
zoologico2 = capturar_datos(2)

# --- EJECUCIÓN AUTOMÁTICA DE FUNCIONES ---

print("\n" + "="*40)
print("  REPORTE DE ACTIVIDADES DIARIAS")
print("="*40)

# Interactuando con el Objeto 1
print(f"\nESTADO DE: {zoologico1.nombre.upper()}")
zoologico1.alimentarAnimales()
zoologico1.guiarTour()
zoologico1.mantenerHabitat()

# Interactuando con el Objeto 2
print(f"\nESTADO DE: {zoologico2.nombre.upper()}")
zoologico2.alimentarAnimales()
zoologico2.guiarTour()
zoologico2.mantenerHabitat()

print("\n" + "="*40)
print("      PROCESO FINALIZADO")
print("  REALIZADO POR DANIEL GALLEGOS 1392")
print("="*40)
```
